# Import libs

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from adjustText import adjust_text
```

# Read csv into df and show data

```python
tat = pd.read_csv("./data/lucas_organic_carbon_training_and_test_data_NEW.csv")
tat
```

```python
targets = pd.read_csv("./data/lucas_organic_carbon_target.csv")
targets
```

# Just Plotting it once

```python
plt.plot(tat)
plt.title("Simple Plot of Data")
plt.xlabel("Row in Table")
plt.ylabel("Value")
plt.show()
```

# Plot of all Values (readable)

```python
tat_transposed = tat.transpose()
tat_transposed
```

```python
plt.plot(tat_transposed, linewidth = 0.5)

n = 200  # Show every nth label because there are to many to create a reasonable plot
x_ticks = range(0, 4000, n)
x_labels = [tat.columns[i] for i in x_ticks]
plt.xticks(x_ticks, x_labels, rotation=45) 

plt.title("Simple Plot of all Values")
plt.xlabel("Wavelength")
plt.ylabel("Value")
plt.show()
```

# Plot of values for one row

```python
row_to_plot = tat.loc[0]
plt.plot(row_to_plot)
plt.title("Plot of Values for first row")

n = 200  # Show every nth label because there are to many to create a reasonable plot
x_ticks = range(0, len(row_to_plot), n)
x_labels = [tat.columns[i] for i in x_ticks]
plt.xticks(x_ticks, x_labels, rotation=45) 

plt.xlabel("Wavelength")
plt.ylabel("Value")
plt.show()
```

# Plot of Targets

```python
sns.countplot(x = 'x', data=targets)
plt.title("Counted values for each class")
plt.xlabel("class of organic soil concentration")
plt.ylabel("count")
plt.show()
```

# Plot of Values of first n rows

```python
number_of_rows = 10
rows_to_plot = tat.head(number_of_rows)
rows_to_plot = rows_to_plot.transpose()
rows_to_plot
```

```python
plt.plot(rows_to_plot, linewidth = 0.5)
plt.title("Plot of Values for first ten rows")

n = 200  # Show every nth label because there are to many to create a reasonable plot
x_ticks = range(0, 4000, n)
x_labels = [tat.columns[i] for i in x_ticks]
plt.xticks(x_ticks, x_labels, rotation=45) 

plt.xlabel("Wavelength")
plt.ylabel("Value")
plt.show()
```
